```python
amp = ipywidgets.widgets.FloatSlider(min=1,max=10,value=1,description="Amp")
phase = ipywidgets.widgets.FloatSlider(min=0,max=5,value=0,description="Phase")
freq = ipywidgets.widgets.FloatSlider(min=1,max=10,value=0,description="Freq")
x = ipywidgets.widgets.FloatRangeSlider(min=0,max=1)

f,ax = plt.subplots(1,figsize=(3,3))
ax.set_ylim([-10,10])
plt.suptitle("Sine Wave")

def update_plot(x,amp,phase,freq):
    ax.clear()
    x1, x2 = x
    xt = np.linspace(x1,x2,50)
    y = amp*np.sin(2*np.pi*(freq*xt+phase))
    ax.plot(xt,y)
    ax.set_ylim([-10,10])
    ax.set_xlim([x1,x2])
    ax.grid()
    plt.show()

ipywidgets.widgets.interactive(update_plot,x=x,amp=amp,phase=phase,freq=freq)
```
